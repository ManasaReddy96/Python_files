```python
import random 

def roll_dice():
    return random.randint(1,6)

def play_game():
    players=['player1','player2']
    positions=[0,0]
    player_index=0
    
    while max(positions)<100:
        roll=roll_dice()
        positions[player_index]+=roll
        if positions[player_index]>100:
            positions[player_index]-=roll
        elif positions[player_index] in snakes:
            positions[player_index]=snakes[positions[player_index]]
        elif positions[player_index] in ladders:
            positions[player_index]=ladders[positions[player_index]]
        print(players[player_index],"rolled a",roll,"and moved to postion",positions[player_index])
        
        if positions[player_index]==100:
            print(players[player_index],"wins!")
            break
            
        player_index=(player_index+1)%2
        
snakes = {16: 6, 47: 26, 49: 11, 56: 53, 62: 19, 64: 60, 87: 24, 93: 73, 95: 75, 98: 78}
ladders = {1: 38, 4: 14, 9: 31, 21: 42, 28: 84, 36: 44, 51: 67, 71: 91, 80: 100}

play_game()
    
        
```
